```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=pd.read_csv("MentalHealthSurvey.csv")
print(data.head(5))
```

```python
data.head()
```

```python
data.info()
```

```python
null_count=data.isnull().sum()
print(null_count)
```

```python

plt.figure(figsize=(6, 3))
sns.histplot(data=data, x='degree_level', hue='gender')

plt.title("Distribution of Gender")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=data, x='university', hue='gender',kde=True)
plt.xticks(rotation=90)
plt.title('University wise report')
plt.show()
```

```python
plt.figure(figsize=(6,3))
data_count = data['university'].value_counts().reset_index()
print(data_count)

sns.barplot(data=data_count, y='count',hue='university')
plt.xticks(rotation=45)
plt.title("distribution of university")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.barplot(data=data, y='academic_year',hue='gender')

plt.title('Year Wise Report Male and Female')
plt.show()
```
